```python
#Importando a biblioteca
import cv2

#Função usada para exibir imagens diretamente no ambiente do Google Colab.
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')

pastaimagem = "/content/drive/MyDrive/Colab Notebooks/testeaula.png"
```

```python
#É uma função da biblioteca OpenCV usada para ler uma imagem de um arquivo.
imagem = cv2.imread(pastaimagem)
```

```python
#Comando da biblioteca OpenCV que exibe a imagem.
cv2_imshow(imagem)
```

```python
#cv2.cvtColor é uma função da biblioteca OpenCV usada para converter as cores de uma imagem, com o código que especifica a cor desejada.
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2_imshow(imagem_cinza)
```

```python
#Definindo um valor de limiar para separar os pixels em preto e branco.
limiar = 127

#Aplicando o limiar para criar a imagem preto e branco.
#O comando cv2.THRESH_BINARY define os pixels que estão acima do limiar como brancos e os abaixo do limiar como pretos.
_, imagem_preto_e_branco = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)

cv2_imshow(imagem_preto_e_branco)
```
